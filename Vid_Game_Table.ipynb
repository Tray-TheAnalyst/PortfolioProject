```python
#import pandas library; call it pd
import pandas as pd
```

```python
#assign Video Game Table data from Wiki webpage to a variable
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_video_games'
```

```python
#scrape/read in the data
#assign that data to a variable
pd.read_html(url)
list_of_dfs = pd.read_html(url)
```

```python
#this is the 1st table of the Wiki webpage but I want the 2nd table; the n tables are stored in a df list called 'list_of_dfs[n]'
list_of_dfs[0]
```

```python
#here's the next table in the list (aka the one I want to analyze)
list_of_dfs[1]
```

```python
#assign the table I want to analyze to a variable
df = list_of_dfs[1]
```

```python
#call the first several rows of desired table
df.head()
```

```python
#show me the number of (rows, columns) the table has
df.shape
```

```python
#call the entire table
df
```

```python
#show me the top 5 Video Games by Sales
# df.nlargest(5, 'Sales'); here I tested out the function before assigning a variable and then calling it
most_VGame_Sales = df.nlargest(5, 'Sales')
most_VGame_Sales
```

```python
#show me the least 5 Video Games by Sales
least_VGame_Sales = df.nsmallest(5, 'Sales')
least_VGame_Sales
```
